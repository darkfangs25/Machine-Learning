# Parkinson Disease Detection

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
```

```python
df = pd.read_csv('parkinsons data.csv')
```

```python
df.info()
```

### Dropping 'name' column

```python
df.drop(columns='name',inplace=True)
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

### Scaling Values to the range (0,1)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
cols_to_scale = ['MDVP:Fo(Hz)','MDVP:Fhi(Hz)','MDVP:Flo(Hz)','MDVP:Jitter(%)','MDVP:Jitter(Abs)','MDVP:RAP','MDVP:PPQ','Jitter:DDP','MDVP:Shimmer','MDVP:Shimmer(dB)','Shimmer:APQ3','Shimmer:APQ5','MDVP:APQ','Shimmer:DDA','NHR','HNR','RPDE','DFA','spread1','spread2','D2','PPE']
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```

```python
def unique_values(df):
    for ele in df:
        print(f'{ele} : {df[ele].unique()} \n')
```

```python
unique_values(df)
```

```python
df.info()
```

### Splitting Training and Testing set

```python
x = df.drop(columns='status',axis='columns')
```

```python
y = df['status']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

### Building Neural Network

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(22,input_shape=(22,),activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp = model.predict(x_test)
```

```python
y_pred = []
for i in yp:
    if i>=0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

### Classification Report and Confusion Matrix of Neural Network

```python
print(classification_report(y_test,y_pred))
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(cm,annot=True)
```

### Building Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr_yp = lr.predict(x_test)
```

```python
lr_yp
```

```python
lr_y_pred = []
for i in lr_yp:
    if i >=0.5:
        lr_y_pred.append(1)
    else:
        lr_y_pred.append(0)
```

```python
lr_y_pred[:10]
```

```python
y_pred[:10]
```

### Classification Report and Confusion Matrix of Linear Regression Model

```python
print(classification_report(y_test,lr_y_pred))
```

```python
lr_cm = tf.math.confusion_matrix(labels=y_test,predictions=lr_y_pred)
```

```python
sns.heatmap(lr_cm,annot=True)
```

```python
comp = {
    'True Value' : y_test,
    'Linear Regression' : lr_y_pred,
    'Neural Network' : y_pred
}

comparison = pd.DataFrame(comp)
```
